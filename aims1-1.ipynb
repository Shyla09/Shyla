```python
import pandas as pd
from google.colab import files

class OneHotEncoder:
    def __init__(self):
        pass

    def fit(self, data):
        self.unique_items = []
        self.categorical_columns = [col for col in data.columns if data[col].dtype == "object"]
        for col in self.categorical_columns:
            self.unique_items.append(data[col].unique().tolist())

    def transform(self, data):
        temp_df = data.copy()
        index = 0
        for col in self.categorical_columns:
            for unique_value in self.unique_items[index]:
                binary_column = []
                for value in temp_df[col]:
                    if value == unique_value:
                        binary_column.append(1)
                    else:
                        binary_column.append(0)
                temp_df[unique_value] = binary_column
            temp_df.drop(col, axis=1, inplace=True)
            index += 1
        return temp_df

    def fit_transform(self, data):
        self.fit(data)
        return self.transform(data)

# Upload the file and read into a DataFrame
uploaded = files.upload()
for filename in uploaded.keys():
    df = pd.read_csv(filename)

encoder = OneHotEncoder()
encoded_df = encoder.fit_transform(df)

# Print the encoded DataFrame
print("Encoded DataFrame:")
print(encoded_df)
```
